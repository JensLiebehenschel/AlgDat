## Data structure max priority queue - implemented as max-heap - with basic operations

## Constants and global variables

```python
# maximal size of the max priority queue
MAX_NUM_ELEMENTS = 10
# implementation as a heap
heap = [-1]*MAX_NUM_ELEMENTS
heapsize = 0
```

### Functions

```python
# navigation in the heap, valid for first element at index 0
def left_child(i):
   return 2*i + 1
def right_child(i):
   return 2*i + 2
def parent(i):
   return (i-1)//2
```

```python
# get maximal element, return False if max priority queue is empty
def extract_max():
   global heapsize
   if heapsize > 0:
      max_elem = heap[0]
      heapsize -= 1
      heap[0] = heap[heapsize]
      max_heapify(0)
      return max_elem
   else:
      return False
```

```python
# function used internally for reestablishing max heap property
def max_heapify(i):
   left = left_child(i)
   right = right_child(i)
   if left < heapsize and heap[left] > heap[i]:
      maximum = left
   else:
      maximum = i
   if right < heapsize and heap[right] > heap[maximum]:
      maximum = right
   if maximum != i:
      heap[i], heap[maximum] = heap[maximum], heap[i]
      max_heapify(maximum)
```

```python
def insert(elem):
   global heapsize
   if heapsize < MAX_NUM_ELEMENTS:
      heap[heapsize] = elem
      i = heapsize
      while (i > 0) and heap[i] > heap[parent(i)]:
         heap[i], heap[parent(i)] = heap[parent(i)], heap[i]
         i = parent(i)
      heapsize += 1
      return True
   else:
      return False
```

```python
# increase priority of element at position i - internal knowledge used: bad practice, but ok for here
def increase_priority(i, prio):
   if i >= 0 and i < heapsize and prio > heap[i]:
      heap[i] = prio
      while (i > 0) and heap[i] > heap[parent(i)]:
         heap[i], heap[parent(i)] = heap[parent(i)], heap[i]
         i = parent(i)
      return True
   else:
      return False
```

### Test data structure

```python
# empty max priority queue
heap = [-1]*MAX_NUM_ELEMENTS
heapsize = 0
# test
elements = [1,2,3,4,5]
for i in elements:
   insert(i)
   print(heap[0:heapsize])
for i in elements:
   print(extract_max())
   print(heap[0:heapsize])
print(extract_max())
```

```python
# empty max priority queue
heap = [-1]*MAX_NUM_ELEMENTS
heapsize = 0
# test
elements = [4,6,2,6,5]
for i in elements:
   insert(i)
   print(heap[0:heapsize])
print(extract_max())
print(heap[0:heapsize])
print(increase_priority(2,7))
print(heap[0:heapsize])
print(extract_max())
print(heap[0:heapsize])
```
