## Sorting an array using bubblesort and output of some statistics

```python
# this code contains counters for outputting statistics
# the counters are incremented after the action to be counted happened
def bubblesort(a, outputStats=True):
    # initialization of counters
    exec_outer_loop = 0
    exec_inner_loop = 0
    comparison_keys = 0
    exchange_keys = 0
    # sorting begins here
    for i in range(len(a)-1):
        exec_outer_loop += 1
        for j in range(len(a)-1,i,-1):
            exec_inner_loop += 1
            # if keys are not in correct order, swap them
            if a[j] < a[j-1]:
                a[j],a[j-1] = a[j-1],a[j]
                exchange_keys += 1
            comparison_keys += 1
    if outputStats:
        # output of counters, "# ..." means "number of ..."
        print(TEXT_SPACING, "# keys in array =", len(a))
        print(TEXT_SPACING, "# executions outer loop =", exec_outer_loop)
        print(TEXT_SPACING, "# executions inner loop =", exec_inner_loop)
        print(TEXT_SPACING, "# key comparisons =", comparison_keys)
        print(TEXT_SPACING, "# key exchanges =", exchange_keys)
    return a, exec_outer_loop, exec_inner_loop, comparison_keys, exchange_keys
```

```python
# this code contains counters for outputting statistics
# the counters are incremented after the action to be counted happened
def bubblesort_with_termination_condition(a, outputStats=True):
    # initialization of counters
    exec_outer_loop = 0
    exec_inner_loop = 0
    comparison_keys = 0
    exchange_keys = 0
    # sorting begins here
    for i in range(len(a)-1):
        exec_outer_loop += 1
        swapped = False
        for j in range(len(a)-1,i,-1):
            exec_inner_loop += 1
            # if keys are not in correct order, swap them
            if a[j] < a[j-1]:
                a[j],a[j-1] = a[j-1],a[j]
                exchange_keys += 1
                swapped = True
            comparison_keys += 1
        # no more swaps necessary, a is sorted
        if not swapped:
            break
    if outputStats:
        # output of counters, "# ..." means "number of ..."
        print(TEXT_SPACING, "# keys in array =", len(a))
        print(TEXT_SPACING, "# executions outer loop =", exec_outer_loop)
        print(TEXT_SPACING, "# executions inner loop =", exec_inner_loop)
        print(TEXT_SPACING, "# key comparisons =", comparison_keys)
        print(TEXT_SPACING, "# key exchanges =", exchange_keys)
    return a, exec_outer_loop, exec_inner_loop, comparison_keys, exchange_keys
```

## Output of statistics to understand the behavior of algorithm

```python
# for visualization we make use of pyplot
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [20, 12]
```

## Constants

```python
# The size of the arrays in tests can be changed here
MIN_ARRAY_SIZE = 1
MAX_ARRAY_SIZE = 20
OFFSET_DIFFS = MAX_ARRAY_SIZE 
```

## Tests

### Bubblesort with and without termination condition

```python
# test sorting sorted arrays
results_without = []
results_with = []
for i in range(MIN_ARRAY_SIZE, MAX_ARRAY_SIZE+1):
    sorted_array = list(range(i))
    # we use a copy of the array here, such that both algorithms get the same input
    results_without.append(bubblesort(sorted_array[:], outputStats=False))
    results_with.append(bubblesort_with_termination_condition(sorted_array, outputStats=False))
# create plot of number of elements and number executions of inner loop and number of key exchanges
plt.plot([len(r[0]) for r in results_without],[r[2] for r in results_without], "bo-") # blue dot
plt.plot([len(r[0]) for r in results_without],[r[4] for r in results_without], "co-") # cyan dot
plt.plot([len(r[0]) for r in results_with],[r[2] for r in results_with], "rx-") # red cross
plt.plot([len(r[0]) for r in results_with],[r[4] for r in results_with], "mx-") # magenta cross
# differences of both versions, raised for better visibiliy of other zero values
plt.plot([len(r[0]) for r in results_with],[r[0][2]-r[1][2]+2*OFFSET_DIFFS for r in zip(results_without,results_with)], "green")
plt.plot([len(r[0]) for r in results_with],[r[0][4]-r[1][4]+OFFSET_DIFFS for r in zip(results_without,results_with)], "brown") 
plt.xticks(range(MIN_ARRAY_SIZE, MAX_ARRAY_SIZE+1))
plt.grid(True)
plt.show()
```
